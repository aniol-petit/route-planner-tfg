### Check no inconsistencies are left

```python
origins = list(df_origin_corrected['original_Origin'])
destinations = list(df_destination_corrected['original_Destination'])
mismatches = []
for destination in destinations:
    if destination in origins:
        lat1, lon1 = df_origin_corrected[df_origin_corrected['original_Origin'] == destination]['lat'].values[0], df_origin_corrected[df_origin_corrected['original_Origin'] == destination]['lon'].values[0]
        lat2, lon2 = df_destination_corrected[df_destination_corrected['original_Destination'] == destination]['lat'].values[0], df_destination_corrected[df_destination_corrected['original_Destination'] == destination]['lon'].values[0]

        if (lat1 != lat2 or lon1 != lon2) and destination not in mismatches:
            mismatches.append(destination)

for origin in origins:
    if origin in destinations:
        # Store lat and lon values with 5 decimal points
        lat1, lon1 = df_origin_corrected[df_origin_corrected['original_Origin'] == origin]['lat'].values[0], df_origin_corrected[df_origin_corrected['original_Origin'] == origin]['lon'].values[0]
        lat2, lon2 = df_destination_corrected[df_destination_corrected['original_Destination'] == origin]['lat'].values[0], df_destination_corrected[df_destination_corrected['original_Destination'] == origin]['lon'].values[0]

        if (lat1 != lat2 or lon1 != lon2) and origin not in mismatches:
            mismatches.append(origin)
```

### Merge both dataframes into a single one

```python
# Merge the two dataframes row by row
df_merged = pd.DataFrame({
    'origin': df_origin_corrected['original_Origin'],
    'lat_origin': df_origin_corrected['lat'],
    'lon_origin': df_origin_corrected['lon'],
    'formatted_origin': df_origin_corrected['formatted'],
    'destination': df_destination_corrected['original_Destination'],
    'lat_destination': df_destination_corrected['lat'],
    'lon_destination': df_destination_corrected['lon'],
    'formatted_destination': df_destination_corrected['formatted'],
    'country': df_origin_corrected['original_Country']
})

df_merged.to_csv("../data/connections_geocoded.csv", index=False)
```

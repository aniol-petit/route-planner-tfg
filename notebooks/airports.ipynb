```python
import pandas as pd
```

```python
# This CSV is semicolon-separated, and some numeric fields use comma as the decimal separator.
# Use sep=';' so commas inside numbers don't break parsing.

def comma_decimal_to_float(x):
    if pd.isna(x):
        return pd.NA
    x = str(x).strip()
    if x == "":
        return pd.NA
    return float(x.replace(",", "."))

df = pd.read_csv(
    "../data/airports.csv",
    encoding="cp1250",
    sep=";",
    converters={
        "latitude_deg": comma_decimal_to_float,
        "longitude_deg": comma_decimal_to_float,
    },
)

df.head()
```

```python
df = df[df["type"].isin(["large_airport", "medium_airport", "small_airport"])]
df = df[df["continent"] == "EU"]
df = df[df["scheduled_service"] == "yes"]
df = df[df["iata_code"].notna()]

df.to_csv("../data/airports_filtered.csv", index=False)

```
